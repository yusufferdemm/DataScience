## Caprazlamalar

```python
import seaborn as sns
from pandas.api.types import CategoricalDtype
diamonds = sns.load_dataset('diamonds')
df = diamonds.copy()
cut_kategoriler = ["Fair","Good","Very Good","Premium","Ideal"]
df.cut = df.cut.astype(CategoricalDtype(categories = cut_kategoriler, ordered = True))
df.head()
```

```python
sns.catplot(x="cut" , y="price" , data=df) # Kategorik Degiskenleri Caprazlamak icin Kullanilan bir fonksiyon
```

```python
# Price Arttikca Kalitenin Yuksek sinif urunlerinin yogunlugu arttigi goruldu
```

```python
sns.barplot(x= "cut",y="price" , hue="color",data=df) # Yeni Bir boyut ekleyelim
```

```python
# price max  18823.00 fakat burada 6000 bin e kadar indirip bir sample kullandi . 
# Yeni Bir boyut(color) eklendiginde , Seaborn High lev. bir kutuphane oldugundan ortalama degerler kullandi

#Yorum : J Coloruna Premium kalitede Fiyati fazla yukari ceken grup olmus diger kalitelerde asiri etkilemesede
#        Premium ve Ideal Siniflari icin Farki bir hayli acmis

# Kahverengi Cubuklar Standart sapmalarini gosteriyor. Bunu dogrulayabilirmiyiz?
```

```python
df.groupby(["cut","color"])["price"].mean() # price ve color a gore grupla ve fiyat ortalamasi yap
```
